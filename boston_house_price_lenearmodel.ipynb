```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# load the dataset
```

```python
df = pd.read_csv("D:/00. Projects/House_Price_Prediction/HousingData.csv")
```

```python
df.keys()
```

```python
df.rename(columns={"MEDV": "target"},inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
x= df.drop("target",axis =1)
```

```python
y=df["target"]
```

```python
# Train test Split

from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
x_train= scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
# Model Training
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression.fit(x_train,y_train)
```

```python
# print the coeffient and the intercept
```

```python
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
regression.get_params()
```

```python
# Prediction with test data
```

```python
reg_pred = regression.predict(x_test)
```

```python
reg_pred
```

```python
# scatter plot for the prediciton
```

```python
plt.scatter(y_test,reg_pred)
```

```python
residuals = (y_test - reg_pred)
```

```python
# Plot this residuals
```

```python
sns.displot(residuals,kind='kde')
```

```python
## scatter plot with respect to prediciton and red=siduals
plt.scatter(reg_pred,residuals)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
print(mean_absolute_error(y_test, reg_pred))
print(mean_squared_error(y_test, reg_pred))
print(np.sqrt(mean_squared_error(y_test, reg_pred)))
```

```python
# r squire and adjusted R squire method
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,reg_pred)
print(score)
```

```python
#display adjusted R-squared
1- (1-score)*(len(y_test)-1)/(len(y_test)-x_test.shape[1]-1)
```
